# 1. Qu es un Algoritmo?

Un **algoritmo** es una secuencia finita y bien definida de instrucciones para resolver un problema especfico o realizar una tarea computacional. Los algoritmos son fundamentales en todos los campos de la computacin y pueden implementarse tanto en computadoras clsicas como cunticas.

En la **computacin clsica**, los algoritmos manipulan bits ($0$ o $1$), ejecutando instrucciones deterministas o probabilsticas mediante compuertas lgicas. Estos algoritmos se ejecutan en dispositivos digitales como CPU o GPU convencionales.

En la **computacin cuntica**, los algoritmos operan sobre **qubits**, que pueden existir en superposiciones de $0$ y $1$, y pueden presentar **entrelazamiento** e **interferencia**. Un **algoritmo cuntico** es un procedimiento computacional paso a paso, en el que cada paso es realizable en una computadora cuntica, generalmente modelada mediante circuitos cunticos.

Si bien las computadoras cunticas pueden simular algoritmos clsicos (dado que cualquier clculo clsico puede reescribirse como un circuito reversible, que los circuitos cunticos pueden implementar utilizando compuertas reversibles universales) [[Lanzagorta, M., 2009](https://books.google.com/books?id=-wkJIuw0YRsC&q=quantum%2520computer%2520equivalent%2520classical%2520computer&pg=PA23)], estamos particularmente interesados en algoritmos cunticos que aprovechen caractersticas exclusivamente cunticas para lograr ventajas computacionales sobre los mtodos clsicos.

<a id="1.1"></a>
## 1.1. Qu hace que un algoritmo cuntico sea diferente?

Los algoritmos cunticos se distinguen por explotar fenmenos de la mecnica cuntica, como:

* **Superposicin**: La capacidad de los qubits de existir en mltiples estados simultneamente.
* **Entrelazamiento**: La fuerte correlacin entre qubits, incluso a distancia.
* **Interferencia**: La capacidad de combinar rutas computacionales de forma constructiva o destructiva.

Al aprovechar estas caractersticas, los algoritmos cunticos pueden resolver ciertos problemas exponencial o cuadrticamente ms rpido que los algoritmos clsicos ms conocidos.

Sin embargo, la computacin cuntica no es universalmente mejor para todos los problemas. El objetivo es identificar problemas especficos donde los algoritmos cunticos ofrecen **ventaja cuntica**.

<a id="1.2"></a>
## 1.2. Ventaja cuntica

La **ventaja cuntica** (anteriormente llamada *~supremaca~ cuntica*) se refiere al punto en el que un dispositivo cuntico puede resolver un problema computacional con mayor eficiencia que cualquier algoritmo clsico conocido, dentro de un plazo de tiempo y recursos viables.

En octubre de 2019, Google afirm haber demostrado la ventaja cuntica con su procesador *Sycamore* mediante el muestreo aleatorio de circuitos, una tarea que, segn argumentaron, tardara miles de aos en ser replicada por una supercomputadora clsica (Arute et al., 2019) (https://www.nature.com/articles/s41586-019-1666-5). Sin embargo, IBM refut esta afirmacin, argumentando que las simulaciones clsicas podran ser mucho ms eficientes que las estimaciones de Google (Pednault et al., 2019) (https://arxiv.org/abs/1910.09534).

Lograr una ventaja cuntica generalmente requiere:

* Un problema computacional bien definido.  
* Un algoritmo cuntico capaz de resolverlo.  
* Un punto de referencia clsico para comparar el rendimiento.  
* Un argumento terico (a menudo basado en la teora de la complejidad) que sostiene que es improbable que los algoritmos clsicos igualen el rendimiento cuntico.  

Algunos problemas propuestos conocidos para demostrar la ventaja cuntica incluyen:

* El [Algoritmo de Shor](https://en.wikipedia.org/wiki/Shor%27s_algorithm) para la factorizacin de enteros.
* El [Muestreo de Bosones](https://en.wikipedia.org/wiki/Boson_sampling) para simular fotones que no interactan.
* El [Muestreo Aleatorio de Circuitos Cunticos](https://en.wikipedia.org/wiki/Cross-entropy_benchmarking) (utilizado por Google en 2019).

Actualmente, estamos en lo que se llama la era **[NISQ](https://en.wikipedia.org/wiki/Noisy_intermediate-scale_quantum_era)** (Noisy Intermediate-Scale Quantum), donde los dispositivos tienen recuentos de qubits limitados y son propensos a errores, pero ya permiten demostraciones experimentales de ventajas cunticas a pequea escala.

## 1.4. Ejemplos de algoritmos cunticos

Los algoritmos cunticos pueden clasificarse segn las principales tcnicas cunticas que explotan:

* Algoritmos basados en la [Transformada Cuntica de Fourier](https://en.wikipedia.org/wiki/Quantum_Fourier_transform) (QFT)

  * [Algoritmo de Deutsch-Jozsa](https://en.wikipedia.org/wiki/Deutsch%E2%80%93Jozsa_algorithm): demuestra una separacin exponencial entre algoritmos cunticos y deterministas clsicos para un problema particular.
  * [Algoritmo de Bernstein-Vazirani](https://en.wikipedia.org/wiki/Bernstein%E2%80%93Vazirani_algorithm): encuentra una cadena oculta con una sola consulta.
  * [Algoritmo de Simon](https://en.wikipedia.org/wiki/Simon%27s_algorithm): ofrece una aceleracin exponencial para un problema de periodicidad oculta.
  * [Estimacin de Fase Cuntica](https://en.wikipedia.org/wiki/Quantum_phase_estimation_algorithm) (QPE): estima eigenvalores de operadores unitarios; componente clave en muchos algoritmos.
  * [Algoritmo de Shor](https://en.wikipedia.org/wiki/Shor%27s_algorithm): factoriza enteros grandes exponencialmente ms rpido que los algoritmos clsicos.

* Algoritmos basados en la [Amplificacin de Amplitud](https://en.wikipedia.org/wiki/Amplitude_amplification)

  * [Algoritmo de Grover](https://en.wikipedia.org/wiki/Grover%27s_algorithm): bsqueda no estructurada cuadrticamente ms rpida.
  * [Algoritmo de Conteo Cuntico](https://en.wikipedia.org/wiki/Quantum_counting_algorithm): estima el nmero de soluciones a un problema de bsqueda.

* Algoritmos basados en [Caminatas Cunticas](https://en.wikipedia.org/wiki/Quantum_walk)

  * [Problema de Distincin de Elementos](https://en.wikipedia.org/wiki/Element_distinctness_problem): encuentra colisiones en funciones.
  * [Problema de Deteccin de Tringulos](https://en.wikipedia.org/wiki/Triangle-free_graph#Triangle_finding_problem): detecta tringulos en grafos.

* Algoritmos Hbridos Cuntico-Clsicos (importantes para dispositivos NISQ)

  * [Algoritmo Cuntico de Optimizacin Aproximada](https://en.wikipedia.org/wiki/Quantum_optimization_algorithms) (QAOA): para problemas de optimizacin combinatoria.
  * [Solucionador Variacional de Eigenvalores Cunticos](https://en.wikipedia.org/wiki/Variational_quantum_eigensolver) (VQE): estima estados fundamentales de sistemas cunticos (importante en qumica y ciencia de materiales).

En esta sesin, primero exploraremos cmo los circuitos cunticos pueden implementar **operaciones lgicas clsicas**, comenzando con una versin cuntica del **Sumador Completo**.

## 2.1. Sumador completo cuntico

Un sumador completo cuntico demuestra cmo las computadoras cunticas pueden replicar algoritmos clsicos respetando principios cunticos como la reversibilidad.

Las computadoras cunticas pueden implementar lgica clsica al incorporarla dentro de circuitos cunticos. La diferencia clave es que, mientras que las compuertas clsicas como AND u OR suelen ser irreversibles, las computadoras cunticas requieren que todas las operaciones sean reversibles. Esto nos obliga a redisear la lgica clsica como lgica reversible, la cual puede luego mapearse directamente a compuertas cunticas.

Al igual que en la electrnica clsica, donde diseamos [sumadores binarios](https://en.wikipedia.org/wiki/Adder_%28electronics%29) (sumadores parciales, sumadores completos, sumadores con acarreo en cascada, etc.), los circuitos cunticos tambin pueden implementar estas operaciones.

Un **sumador completo** clsico toma tres entradas: dos bits a sumar (`A` y `B`) y un bit de acarreo de entrada (`Cin`). Produce dos salidas: el bit de suma (`S`) y el bit de acarreo de salida (`Cout`).

La lgica clsica se describe mediante:


<p style="font-size:10px;font-style:italic;text-align:right"><a href="https://microcontrollerslab.com/cd4008-4-bit-full-adder-ic-pinout-features-example-datasheet/" target="_blank">Imagen obtenida de MicrocontrollersLab</a></p>

Debido a que los circuitos cunticos deben ser reversibles (el mismo nmero de entradas y salidas), diseamos una versin de 4 qubits donde:

* Entradas: `A`, `B`, `CarryIn` y `aux` (ancilla)
* Salidas: `A`, `B`, `Sum` y `CarryOut`

Esto nos permite implementar lgica clsica de forma reversible utilizando **compuertas CNOT** y **Toffoli (CCX)**.


Una posible implementacin en Qiskit se ve as:

````python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit

A = QuantumRegister(1, 'A')
B = QuantumRegister(1, 'B')
CarryIn = QuantumRegister(1, 'CarryIn-Sum')
cero = QuantumRegister(1, 'aux-CarryOut')
adder = QuantumCircuit(A, B, CarryIn, cero)

adder.ccx(0, 1, 3)
adder.cx(0, 1)
adder.ccx(1, 2, 3)
adder.cx(1, 2)
adder.cx(0, 1)

adder.draw('mpl')
````

Notemos que este circuito es completamente reversible:

- Todas las compuertas utilizadas son reversibles (CNOT y Toffoli).  
- No se pierde informacin; todas las entradas pueden recuperarse de las salidas si es necesario.

Esta reversibilidad es esencial para la computacin cuntica, ya que la mecnica cuntica no permite que se destruya informacin durante el clculo.

Ahora podemos inicializar y probar este sumador cuntico completo con entradas especficas.

### Ejemplo 1  Suma `A=1`, `B=0`, `CarryIn=1`

Preparar el estado inicial:

````python
initialization = QuantumCircuit(4)

initialization.x(0)       # A       = 1
initialization.x(2)       # CarryIn = 1
initialization.barrier()

initialization.draw('mpl')
````

Compongamos con el sumador, midamos y ejecutemos la simulacin:

````python
sum = initialization.compose(adder)

sum.barrier()

sum.add_register(ClassicalRegister(2, 'Resultado'))

sum.measure([2, 3], [0, 1])

sum.draw('mpl')
````

````python
from qiskit.providers.basic_provider import BasicSimulator
from qiskit.visualization import plot_histogram

simulator_basic = BasicSimulator()
job_sum = simulator_basic.run(sum)

legend = ["Suma: 1 + 0 + 1"]
plot_histogram([job_sum.result().get_counts()], legend=legend)
````

### Ejemplo 2: Suma `A=1`, `B=1`, `CarryIn=1`

Repetimos el proceso con otro conjunto de entradas:

````python
initialization2 = QuantumCircuit(4)
initialization2.x(0)       # A       = 1
initialization2.x(1)       # B       = 1
initialization2.x(2)       # CarryIn = 1
initialization2.barrier()

sum2 = initialization2.compose(adder)
sum2.barrier()
sum2.add_register(ClassicalRegister(2, 'Result'))
sum2.measure([2, 3], [0, 1])
display(sum2.draw('mpl'))

job_sum2 = simulator_basic.run(sum2, shots=1)   # una sola ejecucin es suficiente

legend = ["Sum: 1 + 1 + 1"]
plot_histogram([job_sum2.result().get_counts()], legend=legend)
````

### 2.1.1. Ventaja cuntica: uso de la superposicin

Tambin podemos aprovechar la superposicin y el entrelazamiento cunticos.

Por ejemplo, preparemos un estado de Bell entre `A` y `B`, y un estado superpuesto para `CarryIn`.

Esto equivale a establecer la entrada en los estados **E1**, **E4**, **E5** y **E8** simultneamente.

````python
initialization3 = QuantumCircuit(4)
initialization3.h(0)        # A y B en estado de Bell
initialization3.cx(0, 1)
initialization3.h(2)        # estado en superposin: ( |0> + |1> ) / sqrt(2)
initialization3.barrier()

sum3 = initialization3.compose(adder)  # el 'adder' es el mismo
sum3.barrier()
sum3.add_register(ClassicalRegister(2, 'Result'))
sum3.measure([2, 3], [0, 1])
display(sum3.draw('mpl'))

job_sum3 = simulator_basic.run(sum3, shots=4000)   # en este caso si necesitamos multiples shots

print(job_sum3.result().get_counts())

legend = ["Conteos del resultado de la suma3"]
plot_histogram([job_sum3.result().get_counts()], legend=legend)
````

### 2.1.2. Alternativa: usar la clase `CDKMRippleCarryAdder` de Qiskit

Qiskit cuenta con clases que facilitan las sumas de qubits, como las clases: [`CDKMRippleCarryAdder`](https://qiskit.org/documentation/_modules/qiskit/circuit/library/arithmetic/adders/cdkm_ripple_carry_adder.html), [`DraperQFTAdder`](https://qiskit.org/documentation/_modules/qiskit/circuit/library/arithmetic/adders/draper_qft_adder.html), [`VBERippleCarryAdder`](https://qiskit.org/documentation/_modules/qiskit/circuit/library/arithmetic/adders/vbe_ripple_carry_adder.html).

A continuacin tenemos un ejemplo que usa la clase `CDKMRippleCarryAdder` para sumar $3+5$:

````python
from qiskit.circuit.library.arithmetic.adders import CDKMRippleCarryAdder
from qiskit.primitives import StatevectorSampler

qiskit_adder = CDKMRippleCarryAdder(3, 'full', 'Sumador completo')

addend1 = QuantumRegister(3, 'a1')  # primer nmero a sumar
addend2 = QuantumRegister(3, 'a2')  # segundo nmero a sumar
aux = QuantumRegister(2, 'aux')     # acarreo de entrada y de salida
cr = ClassicalRegister(4, 'Result') # para almacenar el resultado en bits clsicos

circ = QuantumCircuit(addend1, addend2, aux, cr)

circ.x([addend1[0], addend1[1]])  # 3 en binario: 011
circ.x([addend2[0], addend2[2]])  # 5 en binario: 101

# El orden esperado por la clase 'CDKMRippleCarryAdder' es: carry_in - 1er operando - 2do operando - carry_out
circ.append(qiskit_adder, [aux[0]] + addend1[0:3] + addend2[0:3] + [aux[1]])

circ.measure(addend2[0:3] + [aux[1]], cr) # esta clase escribe el resultado en el 2do operando y en el carry_out
display(circ.draw('mpl'))



sampler = StatevectorSampler()   # porque 'CDKMRippleCarryAdder' no es una compuerta del conjunto base

job_circ = sampler.run([circ])   # en este caso si necesitamos multiples shots

print(job_circ.result()[0].data.Result.get_counts())   # resultado es 8, que en binario es 1000

legend = ["CDKMRippleCarryAdder: 5 + 3"]
plot_histogram([job_circ.result()[0].data.Result.get_counts()], legend=legend)  # notemos el nombre del registro clsico
````

### Reflexin

## 2.2. Protocolo de teleportacin

Alicia quiere enviar informacin cuntica a Bob. Especficamente, supongamos que quiere enviar un estado cuntico, un qubit: $\ket{\psi}=\alpha\ket{0}+\beta\ket{1}$. Esto implica transmitir informacin sobre $\alpha$ y $\beta$ a Bob.

Existe un teorema en la mecnica cuntica que establece que no es posible simplemente hacer una copia exacta de un estado cuntico desconocido. Esto se conoce como el [teorema de la no clonacin](https://en.wikipedia.org/wiki/No-cloning_theorem). Como resultado de esto, podemos ver que Alicia no puede simplemente generar una copia de $\ket{\psi}$ y drsela a Bob. Solo podemos copiar estados clsicos (no superposiciones).

Sin embargo, al aprovechar dos bits clsicos y un par de qubits entrelazados, Alicia puede transferir su estado $\ket{\psi}$ a Bob. A esto se le conoce como teleportacin porque, al final, Bob tendr a $\ket{\psi}$ y Alicia ya no.

$$$$

Para transferir un estado cuntico, Alicia y Bob deben tener ayuda de un tercero, llammoslo Carlos, que se encargue de enviarles un par de qubits entrelazados. Luego, Alicia realiza algunas operaciones en su qubit, enva los resultados a Bob a travs de un *canal de comunicacin clsico*, y Bob luego realiza algunas operaciones en su extremo para recibir el qubit de Alicia.

Entonces, Alicia tendr dos qubits, en uno estar el estado que desea enviarle a Bob, el otro lo utilizar para mantener un estado entrelazado con el nico qubit de Bob.

Veamos los pasos a seguir:

### Paso 1

Todo comienza con la necesidad de enviar un estado cuntico aleatorio, en particular un qubit, es decir, escrito de la forma:

$$\ket{\psi}=\alpha\ket{0}+\beta\ket{1}$$

Quien lo enva, Alicia, no conoce dicho estado a ser enviado.

Creamos los qubits necesarios para llevar a cabo este protocolo:

````python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister

qr_alicia = QuantumRegister(2, name="alicia")  # dos qubits de Alicia
qr_bob = QuantumRegister(1, name="bob")        # un qubit de Bob
crz = ClassicalRegister(1, name="crz")         # 2 bits clasicos
crx = ClassicalRegister(1, name="crx")         # en 2 diferentes registros

circuito_teleportacion = QuantumCircuit(qr_alicia, qr_bob, crz, crx)

circuito_teleportacion.draw('mpl')
````

Su amigo en comn, Carlos, se encarga de preparar un estado entrelazado entre uno de los qubits de Alicia (el qubit 1) y el qubit de Bob. El estado que prepara es un **estado de Bell**. En el lenguaje de circuitos cunticos, la forma de crear un *par de Bell* entre dos qubits es primero transferir uno de ellos a la base $X$ ($\ket{+}$ y $\ket{-}$) usando una compuerta de Hadamard, y luego aplicar una compuerta $CNOT$ en el otro qubit controlado por el que est en la base $X$.

El par entrelazado, usando los qubits $\ket{q_2q_1}=\ket{0_b 0_{a_1}}$ queda como,

$$CNOT_{1,2}\big(I\otimes H\big)\ket{0_b 0_{a_1}}=\frac{1}{\sqrt2}\big(\ket{0_b 0_{a_1}}+\ket{1_b 1_{a_1}}\big)=\ket{e}$$

````python
def crear_estado_bell(circ, a, b):
    circ.h(a)     # poner el qubit a en el estado |+>
    circ.cx(a, b) # CNOT con control el qubit a y objetivo el qubit b
````

````python
crear_estado_bell(circuito_teleportacion, 1, 2) # circuito, qubit a, qubit b

circuito_teleportacion.draw('mpl')
````

Despus de esto, el estado de 3 qubits, incluyendo el estado a ser enviado $\ket{\psi}$, queda como,

$$\begin{aligned}
\ket{e}\otimes\ket{\psi}&=\frac{1}{\sqrt2}\Big(\big(\ket{00}+\ket{11}\big)\otimes \alpha\ket{0} + \big(\ket{00}+\ket{11}\big)\otimes \beta\ket{1}\Big)\\
&=\frac{1}{\sqrt2}\big(\alpha\ket{000}+\alpha\ket{110}+\beta\ket{001}+\beta\ket{111}\big)
\end{aligned}$$

### Paso 2

Ahora, de acuerdo con el protocolo, Alicia aplica la compuerta $CNOT$ en sus dos qubits, usando de control el qubit a ser enviado $\ket{\psi}$, seguido de una compuerta Hadamard en ese mismo qubit. Esto da como resultado el estado:

$$\begin{aligned}
\Big(I\otimes I\otimes H\Big)\Big(I\otimes CNOT_{0,1}\Big)\Big(\ket{e}\otimes\ket{\psi}\Big)&=\Big(I\otimes I\otimes H\Big)\Big(I\otimes CNOT_{0,1}\Big)\frac{1}{\sqrt2}\Big(\alpha\ket{000}+\alpha\ket{110}+\beta\ket{001}+\beta\ket{111}\Big)\\
&=\Big(I\otimes I\otimes H\Big)\frac{1}{\sqrt2}\Big(\alpha\ket{000}+\alpha\ket{110}+\beta\ket{011}+\beta\ket{101}\Big)\\
&=\frac{1}{2}\Big(\alpha\big(\ket{000}+\ket{001}+\ket{110}+\ket{111}\big)+\beta\big(\ket{010}-\ket{011}+\ket{100}-\ket{101}\big)\Big)
\end{aligned}$$

Que puede ser reescrito como:

$$\begin{aligned}
=\frac{1}{2} \Big( & ~~~~~~\big(\alpha\ket{0}+\beta\ket{1}\big)~~\ket{00}\\
&+ ~\big(\alpha\ket{0}-\beta\ket{1}\big)~~\ket{01}\\
&+ ~\big(\alpha\ket{1}+\beta\ket{0}\big)~~\ket{10}\\
&+ ~\big(\alpha\ket{1}-\beta\ket{0}\big)~~\ket{11}~~~~~\Big)\\
\end{aligned}$$

En cdigo queda como sigue:

````python
def compuertas_alicia(circ, psi, a):
    circ.barrier()
    circ.cx(psi, a)
    circ.h(psi)
````

````python
compuertas_alicia(circuito_teleportacion, 0, 1)

circuito_teleportacion.draw('mpl')
````

### Paso 3

Alicia mide los dos primeros qubits (que son de su propiedad) y los enva como dos bits clsicos a Bob. El resultado que Alicia obtiene es siempre uno de los cuatro estados bsicos estndar $\ket{00}$, $\ket{01}$, $\ket{10}$, $\ket{11}$, con igual probabilidad.

Sobre la base de su medicin, el estado de Bob se proyectar a,

$$\ket{00}\rightarrow \big(\alpha\ket{0}+\beta\ket{1}\big)$$
$$\ket{01}\rightarrow \big(\alpha\ket{0}-\beta\ket{1}\big)$$
$$\ket{10}\rightarrow \big(\alpha\ket{1}+\beta\ket{0}\big)$$
$$\ket{11}\rightarrow \big(\alpha\ket{1}-\beta\ket{0}\big)$$

Entonces agregamos las mediciones a los qubits de Alicia:

````python
def mediciones_y_envio(circ, psi, a):
    circ.barrier()
    circ.measure(psi, 0)
    circ.measure(a, 1)
````

````python
mediciones_y_envio(circuito_teleportacion, 0, 1)

circuito_teleportacion.draw('mpl')
````

### Paso 4

Bob, al recibir los bits de Alicia, sabe que puede obtener el estado original aplicando las transformaciones apropiadas en su qubit que una vez fue parte del par de Bell entrelazado.

Las transformaciones que necesita aplicar son:

|                Estado de Bob           | Bits recibidos | Compuerta(s) a aplicar |
|:--------------------------------------:|:------------:|:-------------:|
|$$\big(\alpha\ket{0}+\beta\ket{1}\big)$$|00|$$I$$|
|$$\big(\alpha\ket{0}-\beta\ket{1}\big)$$|01|$$Z$$|
|$$\big(\alpha\ket{1}+\beta\ket{0}\big)$$|10|$$X$$|
|$$\big(\alpha\ket{1}-\beta\ket{0}\big)$$|11|$$ZX$$|

Despus de este paso, Bob habr reconstruido con xito el estado de Alicia.

**Nota**: Esta transferencia de informacin es puramente clsica.

````python
def compuertas_bob(circ, b, crz, crx):
    circ.barrier()
    with circ.if_test((crx, 1)):
        circ.z(2)
    with circ.if_test((crz, 1)):
        circ.x(2)
````

````python
compuertas_bob(circuito_teleportacion, 2, crz, crx)

circuito_teleportacion.draw('mpl')
````

Al final de este protocolo, el qubit de Alicia ahora se ha teleportado a Bob.

Comprobmoslo ejecutando el protocolo completo y simulndolo:

````python
# El protocolo completo:

# Configuracin inicial
qr_alicia = QuantumRegister(2, name="alicia")  # dos qubits de Alicia
qr_bob = QuantumRegister(1, name="bob")        # un qubit de Bob
crz = ClassicalRegister(1, name="crz")         # 2 bits clasicos
crx = ClassicalRegister(1, name="crx")         # en 2 diferentes registros
circuito_teleportacion = QuantumCircuit(qr_alicia, qr_bob, crz, crx)

# Paso 0
# Iniciar el qubit a enviar
circuito_teleportacion.append(instruccion_inicializacion, [qr_alicia[0]])
circuito_teleportacion.barrier()

# Paso 1
# Crear el par entrelazado
crear_estado_bell(circuito_teleportacion, qr_alicia[1], qr_bob[0]) # circuito, qubit a, qubit b

# Paso 2
# "Enviar" q1 a Alicia y q2 a Bob
compuertas_alicia(circuito_teleportacion, qr_alicia[0], qr_alicia[1])

# Paso 3
# Alicia envia sus bits clsicos a Bob
mediciones_y_envio(circuito_teleportacion, qr_alicia[0], qr_alicia[1])

# Paso 4
# Bob decodifica su qubit
compuertas_bob(circuito_teleportacion, qr_bob[0], crz, crx)

circuito_teleportacion.draw('mpl')
````

## 2.C. Protocolo de codificacin superdensa

### Diferencia entre la Codificacin superdensa y la Teleportacin cuntica

La teleportacin cuntica y la codificacin superdensa estn estrechamente relacionadas, para evitar confusiones, debemos aclarar la diferencia.

La teleportacin cuntica es un proceso mediante el cual el estado de un qubit ($\ket{\psi}$) se puede transmitir de un lugar a otro, utilizando dos bits de comunicacin clsica y un par de Bell. En otras palabras, podemos decir que es un protocolo que destruye el estado cuntico de un qubit en una ubicacin y lo recrea en un qubit en una ubicacin distante, con la ayuda del entrelazamiento. La **codificacin superdensa** es un procedimiento que le permite a alguien enviar dos bits clsicos a otra parte utilizando solo un qubit de comunicacin.

|           Teleportacin       |     Codificacin superdensa     |
|:--------------------------------------:|:------------:|
| Transmite un qubit <br> usando dos bits <br> clsicos | Transmite dos bits <br> clsicos usando <br> un qubit |


El protocolo de teleportacin se puede considerar como una versin invertida del protocolo de codificacin superdensa, en el sentido de que Alicia y Bob simplemente "intercambian su equipo".

### Paso 1

El proceso comienza con una tercera persona, a quien llamaremos Carlos, quien prepara dos qubits en un estado entrelazado. Inicialmente inicia los 2 qubits en el estado base $\ket{0}$. Aplica aplica la compuerta de Hadamard ($H$) al primer qubit para crear una superposicin. Luego aplica la compuerta $CNOT$ ($CX$) usando el primer qubit como control y el segundo como objetivo. Este es el estado entrelazado (par de Bell) que mencionamos anteriormente.

Comenzamos con el estado:

$$\ket{00}=\ket{0}_A\otimes\ket{0}_B$$

donde el qubit que se enviar a Alicia est etiquetado con $A$ y el qubit que se enviar a Bob est etiquetado como $B$. Carlos primero aplica una compuerta de Hadamard al primer qubit, lo que crea una superposicin y obtenemos el estado:

$$\ket{+0}=\frac{1}{\sqrt2}\Big(\ket{00}+\ket{10}\Big)$$

Entonces Carlos aplica la compuerta $CNOT$, esta entrelaza ambos qubits, es decir, invierte el qubit objetivo si el control es $\ket{1}$. Ten en cuenta que el qubit de control es nuestro qubit ms a la izquierda.

$$CNOT\frac{1}{\sqrt2}\Big(\ket{00}+\ket{10}\Big)=\frac{1}{\sqrt2}\Big(\ket{00}+\ket{11}\Big)$$

````python
def crear_par_de_bell():
    qc = QuantumCircuit(2)
    qc.h(1)
    qc.cx(1, 0)
    return qc
````

### Paso 2

Carlos enva el primer qubit a Alicia y el segundo a Bob. El objetivo del protocolo es que Alicia enve 2 bits clsicos de informacin a Bob usando su qubit. Pero antes de hacerlo, necesita aplicar un conjunto de compuertas cunticas a su qubit dependiendo de los 2 bits de informacin que quiera enviar:

| Mensaje a enviar | Compuerta a aplicar | Estado resultante $\Big(\cdot\frac{1}{\sqrt2}\Big)$ |
|:--------------------------------------:|:------------:|:-------------:|
|00|$$I$$|$$\ket{00}+\ket{11}$$
|01|$$X$$|$$\ket{10}+\ket{01}$$
|10|$$Z$$|$$\ket{00}-\ket{11}$$
|11|$$ZX$$|$$-\ket{10}+\ket{01}$$

Por lo tanto, si Alicia quiere enviar un $00$, no le hace nada al qubit (aplica la compuerta de identidad ($I$)). Si quiere enviar un $01$, entonces aplica la compuerta $X$. Dependiendo de lo que quiera enviar, aplica la compuerta adecuada y luego enva su qubit a Bob para el paso final del proceso.

````python
def codificar_mensaje(qc, qubit, msg):
    if len(msg) != 2 or not set(msg).issubset({"0", "1"}): # el mensaje debe ser de solo 0s y 1s
        raise ValueError(f"El mensaje '{msg}' es invlido")
    if msg[1] == "1":
        qc.x(qubit)
    if msg[0] == "1":
        qc.z(qubit)
    return qc
````

### Paso 3

Bob recibe el qubit de Alicia (el qubit ms a la izquierda) y usa su qubit para decodificar el mensaje de Alicia. Ten en cuenta que no necesita tener conocimiento del estado para decodificarlo, simplemente usa la operacin de restauracin.

Bob aplica una compuerta CNOT usando el qubit ms a la izquierda como control y el ms a la derecha como objetivo. Luego aplica una compuerta Hadamard y finalmente realiza una medicin en ambos qubits para extraer el mensaje de Alicia.

| Bob recibe $\Big(\cdot\frac{1}{\sqrt2}\Big)$ | Despus de la compuerta $CNOT$ | Despus de la compuerta $H$ |
|:--------------------------------------:|:------------:|:-------------:|
|$$\ket{00}+\ket{11}$$|$$\ket{00}+\ket{10}$$|$$\ket{00}$$
|$$\ket{10}+\ket{01}$$|$$\ket{11}+\ket{01}$$|$$\ket{01}$$
|$$\ket{00}-\ket{11}$$|$$\ket{00}-\ket{10}$$|$$\ket{10}$$
|$$-\ket{10}+\ket{01}$$|$$-\ket{11}+\ket{01}$$|$$\ket{11}$$

````python
def decodificar_mensaje(qc):
    qc.cx(1, 0)
    qc.h(1)
    return qc
````

````python
# El protcolo completo

# Carlos crea un par entrelazado entre Alicia y Bob
qc = crear_par_de_bell()

# Agregamos una barrera para una separacin visual
qc.barrier()

# En este punto, el qubit 0 va a Alicia y el qubit 1 va a Bob

# Luego, Alicia codifica su mensaje en el qubit 1. En este caso, 
# queremos enviar el mensaje '10'. Puedes intentar cambiar este
# valor y ver cmo afecta al circuito
mensaje = '10'
qc = codificar_mensaje(qc, 1, mensaje)
qc.barrier()
# Luego Alicia enva su qubit a Bob

# Despus de recibir el qubit 0, Bob aplica el protocolo de recuperacin:
qc = decodificar_mensaje(qc)

# Finalmente, Bob mide sus qubits para leer el mensaje de Alicia
qc.measure_all()

# Mostramos el circuiti
qc.draw('mpl')
````

````python
from qiskit.primitives import StatevectorSampler as Sampler

sampler = Sampler()
job_qc = sampler.run([(qc, None, 1000)]) # usamos multiples shot

counts = job_qc.result()[0].data.meas.get_counts() 

print(counts)

legend = ["Mensaje enviado"]
plot_histogram([counts], legend=legend)
````
